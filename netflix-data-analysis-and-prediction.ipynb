```python
#Predicting whether the customer will leave the netflix or not
x = df[["age", "gender", "monthly_fee", "last_login_days", "number_of_profiles", "avg_watch_time_per_day","payment_method"]]
y = df["churned"]
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.3, random_state= 1)

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
num_cols = [col for col in x.columns if x[col].dtypes != "object"]
cat_cols = [col for col in x.columns if x[col].dtypes == "object"]

num_pipline = Pipeline([("SimpleImputer", SimpleImputer(strategy= "mean")), ("StandardScaler", StandardScaler())])
cat_pipline = Pipeline([("SimpleImputer", SimpleImputer(strategy= "most_frequent")), ("OneHotEncoder", OneHotEncoder(handle_unknown="ignore"))])
preprocessor = ColumnTransformer([("Numerical_pipeline", num_pipline, num_cols),
                   ("Categorical_pipeline", cat_pipline, cat_cols)])
x_train = preprocessor.fit_transform(x_train)
x_test = preprocessor.transform(x_test)


from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
from sklearn.metrics import accuracy_score

print("Logistic Regression Accuracy:",accuracy_score(y_test, y_pred))
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=200, random_state=42)

rf.fit(x_train, y_train)

rf_pred = rf.predict(x_test)

from sklearn.metrics import accuracy_score

print("Random Forest Accuracy:", accuracy_score(y_test, rf_pred))


from xgboost import XGBClassifier

xgb = XGBClassifier(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=6,
    random_state=42
)

xgb.fit(x_train, y_train)

xgb_pred = xgb.predict(x_test)

print("XGBoost Accuracy:", accuracy_score(y_test, xgb_pred))
```

```python
import joblib

joblib.dump(xgb, "xgb_model.pkl")
joblib.dump(preprocessor, "preprocessor.pkl")

print("Saved successfully")
```
